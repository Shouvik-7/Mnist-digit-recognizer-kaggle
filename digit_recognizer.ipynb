```python
import pandas as pd
data_df = pd.read_csv("train.csv")
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
data_df.head(10)
```

```python
y_train = np.array(data_df["label"])
y_train = y_train.astype(np.uint8)
X_train = np.array(data_df.iloc[:,1:])
print(X_train.shape)
print(y_train.shape)
```

```python

#X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]
#print(X_train.shape)
#print(y_train.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_jobs=-1, weights='distance', n_neighbors=4)
knn_clf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score 
cross_val_score(knn_clf, X_train, y_train, cv=3, scoring="accuracy")
```

```python
n_neighbors = list(range(2,6))
weights = ['distance','uniform']
param = dict(n_neighbors=n_neighbors,weights = weights)
knn_2 = KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(knn_2,param,cv=3,verbose=3, n_jobs=-1)
model = clf.fit(X_train,y_train)

```

```python
#print('Best n_neighbors:',model.best_estimator_.get_params()['n_neighbors'])

#print('Best weights:',model.best_estimator_.get_params()['weights'])
model.best_params_
```

```python
X_test = np.array(pd.read_csv("test.csv"))
print(X_test.shape)
```

```python
y_pred = model.predict(X_test)
```

```python
df_out=pd.DataFrame({'Label':y_pred})
```
